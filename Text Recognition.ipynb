```python
import os

os.environ['KMP_DUPLICATE_LIB_OK']='True'

import matplotlib.pyplot as plt
import pytesseract 
import easyocr
import cv2
import matplotlib.image as mpimg
%matplotlib inline

pytesseract.pytesseract.tesseract_cmd = 'C:\Program Files\Tesseract-OCR\\tesseract.exe'
```

Below is the image on which we will try pytesseract first and than easyocr

```python
img = mpimg.imread('Assets/image1.jfif')
imgplot = plt.imshow(img)
plt.show()
```

```python
image=cv2.imread('Assets/image1.jfif')
text=pytesseract.image_to_string(image)
print(text)
```

Here we can see that pytesseract's performance diminishes when there is inconsistent background

Let us try using easyocr 

```python
reader=easyocr.Reader(['en'])
```

We have initialized Reader object of easyocr class with the language as English

readtext method returns list of tuples with the text along with the coordinates of its bounding rectangle

```python
bounds=reader.readtext(img)
print(bounds)
```

The snippet below returns only the text extracted from the image as a string

```python
str_value=''
for x in bounds:
    str_value+=x[1]+" "
str_value.strip()
```
